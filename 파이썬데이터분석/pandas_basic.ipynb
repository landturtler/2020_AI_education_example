### Pandas 기본
#### (1) DataFrame 객체 : 2차원 배열, 서로 다른 자료형을 사용할 수 있다, 행:인스턴스, 열:피쳐(feature)
#### (2) Series 객체 : 1차원 배열, 1차원 ndarray와 호환

#### [1] 데이터 프레임(DataFrame) 객체

```python
# pip install pandas
import pandas as pd
import numpy as np

# 2차원 리스트, 배열을 사용하여 생성
df = pd.DataFrame([[1,2,3],
                   [4,5,6],
                   [7,8,9]])
print(type(df))
print(df)
df
```

```python
# 사전(dict)을 사용하여 생성
score_table = { '성명': ['홍길동','김철수','김영희','나','너'],
                '국어': [50,60,70,80,90],
                '영어': [10,20,30,40,50],
                '수학': [70,80,90,30,20] }
df = pd.DataFrame(score_table)
print(type(df))
print(df)
df 
```
