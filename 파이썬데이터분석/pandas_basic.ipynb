### DataFrame의 요소 접근하기 : 읽기와 수정
* (1) df.loc[행번호, 열] : 명시적인 index
* (2) df.iloc[행번호, 열] : 함묵적인 index, 파이썬 스타일 , 인덱싱, int형 0부터 시작

```python
# df.loc[]
df.loc[3,'Country']
df.loc[:3,'Country']   #  start:end, 주의  end - 1 까지가 아님
df.loc[:3,['Country','CountryID','Continent']]   #  start:end, 주의  end - 1 까지가 아님
```

```python
# df.iloc[] : index를 int형으로 사용  , 자주 사용됨!!

df.iloc[3,0]
df.iloc[:4,0:3]       # start:end + 1
df.iloc[:4,[0,2,4]]

# 스칼라 요소값의 변경
df.iloc[3,0] = 'Dorra'

df.iloc[-1,0]
df.iloc[0,-1]
```
