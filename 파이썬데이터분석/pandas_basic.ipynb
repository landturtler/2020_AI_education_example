### 데이터 프레임 추가와 합치기

```python
# 행(수직)으로 합치기 : append(), concat()
score_table1 = {'영어':[10,20,30,40,50],
                '수학':[70,80,90,30,20]}

score_table2 = {'영어':[20,30,40,50,60],
                '수학':[80,90,30,20,70]}

df1 = pd.DataFrame(score_table1)
df2 = pd.DataFrame(score_table2)

new_df1 = df1.append(df2)  # 행의 인덱스가 중복 출력
# print(new_df1)
new_df2 = df1.append(df2,ignore_index=True)  # 행의 인덱스가 중복 되지 않음
# print(new_df2)

new_df3 = df1.append(df1,ignore_index=True)  # 행의 인덱스가 중복 되지 않음
new_df3
```

```python
pd.concat([df1,df2],axis = 0,ignore_index=True)
```

```python
# 열(수평)로 합치기 : merge(), join(), concat()

# concat()
pd.concat([df1,df2],axis = 1)
```

```python
# merge()
dests = pd.read_csv('dest.csv')
tips =  pd.read_csv('tips.csv')
print(dests)
print(tips)

# 내부(inner) 조인 : 공통 조건에 만족되는 부분만 조인된다
# 공통조건 : 'EmpNr'
pd.merge(dests,tips, how='inner', on='EmpNr')
```

```python
# 외부(outer) 조인 : 공통 조건에 관계 없이 조인된다, 결측치는 NaN으로 표시
pd.merge(dests,tips, how='outer')
```

```python
# join(other, on=None, how='left', lsuffix='', rsuffix='', sort=False)
dests.join(tips,lsuffix='_Dest',rsuffix='_Tips')  # left join, 모두 조인됨
```

```python
# groupby()
from numpy.random import seed
from numpy.random import rand
from numpy.random import randint

seed(42)

df = pd.DataFrame({ 'Weather' : ['cold', 'hot', 'cold', 'hot', 'cold', 'hot', 'cold'],
                    'Food' : ['soup', 'soup', 'icecream', 'chocolate', 'icecream', 'icecream', 'soup'],
                    'Price' : 10 * rand(7),
                    'Number' : randint(1, 9, 7)})
df
```

```python
weather_group = df.groupby('Weather')
print(weather_group)

print(weather_group.first())   # cold 의 첫번째 행 ,hot의 첫번째 행
print(weather_group.last())    # cold 의 마지막 행 ,hot의 마지막 행
print('-'*40)
print(weather_group.mean())     # cold 와 hot 각각의 평균
print('-'*40)
print(weather_group.count())     # cold 와 hot 각각의 총 갯수
print('-'*40)

print(weather_group.describe()) # cold 와 hot 각각의 통계 요약
print('-'*40)

i = 0
for name, group in weather_group:  # name은 그룹이름
    i += 1          # i = i + 1  --> i += 1
    print(i,name,'\n', group)
    print(type(group))
```
